```
data['clean_doc']= data['content'].str.replace("[^ㄱ-ㅣ가-힣]", " ")
data.head()
```

```
data['clean_doc'] = data['clean_doc'].apply(lambda x : ' '.join([w for w in x.split() if len(w)> 3]))
data.head()
```

```
stopPos = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JC','JX',
           'SF','SE','SS','SP','SSO','SSC','SC','SY','SH','SL','SN',
           'MM','MAG','MAJ']

tokenized_doc = data['clean_doc'].apply(lambda x : x.split()) 
```

```
def token(apply_split = True):
  if apply_split:
    return tokenized_doc.apply(lambda x : [item for item in x if item not in stopPos])
    
  else :
    return tokenized_doc.apply(lambda x : ' '.join([item for item in x if item not in stopPos]))
```

```
tokenized_doc
```

```
def my_tokenizer(text):
  return text.split()

tokenized_docs = token(False)
```

```
tokenized_docs
```

```
type(tokenized_docs)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

tfidf_vect = TfidfVectorizer(tokenizer = my_tokenizer)
tfidf = tfidf_vect.fit_transform(tokenized_docs)
lda = LatentDirichletAllocation(n_components = 20, max_iter = 20)
lda_output = lda.fit_transform(tfidf)
```
